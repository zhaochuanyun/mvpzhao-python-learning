```python
import matplotlib.pyplot as plt
from PIL import Image
from numpy import *
from scipy.ndimage import filters
```

```python
im_origin = array(Image.open('../../data/curry.jpg'))
im_l = array(Image.open('../../data/curry.jpg').convert('L'))

imx = zeros(im_l.shape)
filters.sobel(im_l,1,imx)
imy = zeros(im_l.shape)
filters.sobel(im_l,0,imy)
magnitude = sqrt(imx**2+imy**2)
```

```python
plt.subplots(2, 2, figsize=(10, 10), sharey=True)

plt.subplot(221)
plt.imshow(im_origin)
plt.axis('off')

plt.subplot(222)
plt.imshow(im_l)
plt.axis('off')

plt.subplot(223)
plt.imshow(imx)
plt.axis('off')

plt.subplot(224)
plt.imshow(imy)
plt.axis('off')
```
